```python
import numpy as np
```

```python
x = np.linspace(0,5,11) #11 elementos de 0 a 5, igualmente espacados
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
# Funcional
```

```python
plt.plot(x,y, 'r--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico')
#plt.show() nao e necessario
```

```python
#METODOS FUNCIONAIS != DE ORIENTADO A OBJ
#OS DOIS PRIMEIROS PARAMETROS E A QTD DE GRAFICOS, (LINHA,COLUNA), O TERCEIRO É QUAL GRAFICO VAI SER MANIPULADO
plt.subplot(1,3,1)
plt.plot(x,y, 'r--')
plt.subplot(1,3,2)
plt.plot(y,x, 'y*-')
plt.subplot(1,3,3)
plt.plot(y/2,x/5, 'b')
```

```python
fig = plt.figure() #cria uma "caixa vazia"
```

```python
#ORIENTADO A OBJETOS

fig = plt.figure()
#10% de distancia da borda lateral esquerda e da inferior
axes1 = fig.add_axes([0.1, 0.1, 1, 1]) #somente com um gráfico nao e perceptivel
axes1.plot(x,y)
axes1.set_xlabel('Axes 1')
axes1.set_title('Gráfico 1')

axes2 = fig.add_axes([0.2, 0.6, .4, .4]) #somente com um gráfico nao e perceptivel
axes2.plot(y,x, 'red')
```

### **Utilizando o método ''subplots'', orientado a obj**

```python
fig, ax = plt.subplots() #cria uma figura e um eixo
ax.plot(x,x**3,'r')
ax.plot(x,x**4,'b--')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2) #a fig seria os containers e o ax seria o obj q eu uso para
#manipular os graficos
```

```python
ax #retorna uma lista com a qtd de eixos que eu gerei
```

```python
for axis in ax:
    axis.plot(x,y, 'b--')
    axis.set_title('title') #para mostrar, eu preciso chamar a 'fig'
fig
```

```python
#cria um grafico em cima do que ja foi plotado, caso ja exista
ax[0].plot(y,x, 'red')
ax[0].set_title('indexing')

ax[1].plot(x,y, 'orange')
ax[1].set_title('indexing')
fig
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout() #ajeita a sobreposicao
```

```python
fig = plt.figure(figsize=(8,4), dpi=100) #cria uma figura vazia, porem sem nenhum eixo
```

```python
fig, axes = plt.subplots(figsize=(8,8))
axes.plot(x,y, 'r*-', label='dotted plot')
axes.plot(y,x, 'g',label='line plot')
axes.set_title('titulo')
axes.legend(loc=0) #range(0,11) loc / 0 - best loc
```

```python
fig.savefig('fig.jpg')
```

```python
fig, ax = plt.subplots(figsize=(8,10)) 
ax.plot(x, x**2, lw=20, alpha=1, linestyle=':',color='pink') #linewidth, alpha = opacity
ax.plot(x, x**3, 'b--')

ax.set_xlim([0,2])
ax.set_ylim([0,10])
```
