# Plots categóricos

Agora vamos discutir como usar seaborn para traçar dados categóricos. Existem alguns tipos de argumentos principais para isso:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

Vamos passar por exemplos de cada um.

Dados que envolvem strings, como por exemplo "sex,smoker,day", são os dados que se encaixam em dados categóricos

```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
#style
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
#especifica que a media para homens foi 21, e a barrinha preta é o desvio padrao
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
#percentil de dados
#ele marca de 25 em 25 % a cada barra
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, palette='rainbow', hue='sex')
#percentil de dados
#ele marca de 25 em 25 % a cada barra
```

```python
sns.boxplot(data=tips, orient='h')
#ele plota um boxplot com os dados nao-categoricos (nesse caso n faz mto sentido os dados)
```

```python
#mostra onde esta distribuido os dados, e mostra a media dos dados
sns.violinplot(x='day', y='tip', data=tips)
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True)
#split junta os dados divididos, ja que é um espelho
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
#nao e mto bom pra qnd tem uma qtd mto grande de pontos, perde um pouco a nocao
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
#conforme vai aumentando os dados, perde um pouco a nocao tbm
#solucao: combinar os plots
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='Black')
#assim e possivel ter a nocao de como a distribuicao foi criada
```
