```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("seaborn")
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head(20)
```

```python
iris.info()
```

```python
iris['species'].value_counts()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter) #map mapeia todos os conjuntos de dados aplicando a mesma funcao
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(data=iris, hue='species')
```
