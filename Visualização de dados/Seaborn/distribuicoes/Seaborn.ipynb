# **Plots de distribuiçoes**
Alguns gráficos que nos permitem visualizar a distribuição de um conjunto de dados. Esses plots são:

* distplot
* joinplot
* pairplot
* rugplot
* kdeplot

```python
import seaborn as sns
plt.style.use("seaborn")
```

```python
# Import Matplotlib
import matplotlib.pyplot as plt

# Check the available styles
plt.style.available
# Use Matplotlib defaults
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info() #gorjeta
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False) #distplot = histograma #kde = kernel density estimation
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
#podemos ver que existe uma relação linear entre x e y
#quanto maior a compra, maior a gorjeta
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg',ylim=(-2,10.5))
#podemos ver que existe uma relação linear entre x e y
#quanto maior a compra, maior a gorjeta
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex',ylim=(-2,10.5)) #mapa de calor
#cores mais fortes onde possui um conjunto de dados maior
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm') #pega todas as variaveis numéricas do dataframe
```

```python
sns.rugplot(tips['total_bill']) #faz mais sentido com o kde, para vizualizar a concentracao dos dados
```

```python
#kde serve para dados quando vc nao tem certeza do tipo de distrubuicao que ele segue
#ele cria uma distribuicao normal em volta de cada ponto, e por fim soma essas distrubuicoes
#para obter a curva final
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
