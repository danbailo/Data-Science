```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
#construido sobre o scatterplot
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers='o v'.split(), col='sex', row='time') #col separa em colunas
```

```python
#construido sobre o scatterplot
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers='o v'.split(), col='day', aspect=0.6, height=8) #col separa em colunas
```
